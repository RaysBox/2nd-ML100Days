### TIP
使用 .plot() 這個方法即可進行畫圖的動作<br>
裡頭丟的參數決定圖的呈現狀態<br>
* kind='bar' >> 長條圖
* kind='barh' >> 橫條圖

```python
# value_counts() 可對數據進行統計並排序, 默認是由大到小, 進行降序排列
app_train['AMT_INCOME_TOTAL'][:20].value_counts().plot(kind='bar')
```

```python
# 顯示前20筆的長條圖
app_train['AMT_INCOME_TOTAL'][:20].plot(kind='bar')
```

```python
app_train[:20].plot(kind='barh')
```
