## 截取部分資料
### 可以用 .head() 這個函數來觀察前 5 row 資料

```python
# 預設前五筆
app_train.head()
```

```python
# 可自訂要看幾筆, 例如15筆
app_train.head(15)
```

```python
# 直接從讀取到的陣列取得前10筆資料
# app_train[:10] == app_train[0:10]
app_train[:10]
```

```python
# 直接從讀取到的陣列查看 index 5 到 index 19 的資料
app_train[5:20]
```

```python
# 取得指定行列數資料的用法
for (name, series) in app_train.iteritems():
    print('Col name: ' + str(name))
    print('First value: ' + str(series.iat[0]) + '\n')
    print('Second value: ' + str(series.iat[1]) + '\n')
    print('Third value: ' + str(series.iat[2]) + '\n')
    print('Fourth value: ' + str(series.iat[3]) + '\n')
```

```python
# 小示範
# 取得欄位名稱為 Unnamed: 1 的 index 為 1 的資料
for (name, series) in app_train.iteritems():
    if 'Unnamed: 1' == name:
        print('Unnamed: 1 value: ' + str(series.iat[1]) + '\n')
```

```python
# 更聰明的小示範
# 取得欄位名稱為 Unnamed: 1 的 index 為 1 的資料
print('Unnamed: 1 value: ' + app_train['Unnamed: 1'][1] + '\n')
```
